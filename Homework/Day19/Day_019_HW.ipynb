```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = 'C:/Users/johnson/Documents/GitHub/ML100Days/Homework/data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
print(app_train['HOUSETYPE_MODE'].unique())
print(app_train['HOUSETYPE_MODE'].value_counts())
```

```python
app_train['HOUSETYPE_MODE'] = app_train['HOUSETYPE_MODE'].fillna('0')
```

```python
unique_house_type = app_train['HOUSETYPE_MODE'].unique()
# len(unique_house_type)
unique_house_type
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
unique_house_type = app_train['HOUSETYPE_MODE'].unique()

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    app_train.loc[ app_train['HOUSETYPE_MODE']==unique_house_type[i], 'AMT_CREDIT'].hist()
    plt.title(str(unique_house_type[i]))
    plt.xticks(rotation =55)
plt.show()    
```

```python
plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
#     app_train.loc[ app_train['HOUSETYPE_MODE']==unique_house_type[i], 'AMT_CREDIT'].hist()
    sns.distplot(app_train.loc[ app_train['HOUSETYPE_MODE']==unique_house_type[i], 'AMT_CREDIT'], label=unique_house_type[i])
    plt.title(str(unique_house_type[i]))
    plt.xticks(rotation =55)
plt.show()    
```

```python
grouped_house = app_train.groupby('HOUSETYPE_MODE')


grouped_house.boxplot(column='AMT_CREDIT', by='HOUSETYPE_MODE', showfliers = False, figsize=(12,12))
```
